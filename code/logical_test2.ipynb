```python

"""
Convert Arabic Number to Roman Number.
เขียนโปรแกรมรับค่าจาก user เพื่อแปลง input ของ user ที่เป็นตัวเลขอราบิก เป็นตัวเลขโรมัน
โดยที่ค่าที่รับต้องมีค่ามากกว่า 0 จนถึง 1000

*** อนุญาตให้ใช้แค่ตัวแปรพื้นฐาน, built-in methods ของตัวแปรและ function พื้นฐานของ Python เท่านั้น
ห้ามใช้ Library อื่น ๆ ที่ต้อง import ในการทำงาน(ยกเว้น ใช้เพื่อการ test การทำงานของฟังก์ชัน).

"""
```

```python
#----------------------setting dictionary-----------------------------
representing = {0:"",1:"I",5:"V",10:"X",50:"L",100:"C",500:"D",1000:"M"}




#------------------------------------function---------------------------------
def insertC(local_num):
    local_text=[]
    BB=[1000,500,100]
    for num in BB:
        result = local_num // num
        if result>=1:
            local_num=local_num-num
            local_text.append(representing[num])
            if local_num>=100:
                local_digits = [int(local_digits) for local_digits in str(local_num)]#split number
                while local_digits[0]>0:
                    local_digits[0]-=1
                    local_text.append(representing[100])
                break
    return local_text




def insertX(local_num):
    local_text=[]
    BB=[1000,500,100,50,10]
    for num in BB:
        result = local_num // num
        if result>=1:
            local_num=local_num-num
            local_text.append(representing[num])
            if local_num>=10:
                local_digits = [int(local_digits) for local_digits in str(local_num)]#split number
                while local_digits[0]>0:
                    local_digits[0]-=1
                    local_text.append(representing[10])
                break
    return local_text





def insertI(local_num):
    local_text=[]
    BB=[1000,500,100,50,10,5,1]
    result_list = []
    for num in BB:
        result = local_num // num
        if result>=1:
            local_num=local_num-num
            local_text.append(representing[num])
            break
    while local_num>0:
        local_num-=1
        local_text.append(representing[1])
    return local_text


def romantext(input_number):
    text=[]
    if input_number<=1000:#case
        digits = [int(digit) for digit in str(input_number)]#split number
        if input_number in representing:
            text.append(representing[input_number])

        else:
            j=0
            for i in range(len(str(input_number)),0,-1):  #122=>3loops(3,2,1)
                if i==3:#รอบเลขหลักร้อย,digits[j]=ตัวเลขหลักร้อย
                    if digits[j]==4 or digits[j]==9:#400...,900...
                        text.append(representing[100])
                        text.append(representing[int(str(digits[j]+1)+'00')])
                    else:
                        text1=insertC(input_number)
                        text1= ''.join(text1)
                        text.append(text1)


                if i==2:#รอบเลขหลักสิบ,digits[j]=ตัวเลขหลักสิบ
                    if digits[j]==4 or digits[j]==9:#40...,90...
                        text.append(representing[10])
                        text.append(representing[int(str(digits[j]+1)+'0')])
                    elif digits[j]==0:
                        pass
                    else:
                        text1=insertX(int(str(digits[j])+'0'))
                        text1= ''.join(text1)
                        text.append(text1)



                if i==1:#เลขหลักหน่วย

                    if digits[j]==4 or digits[j]==9:
                        text.append(representing[1])
                        text.append(representing[digits[j]+1])

                        
                    else:
                        text1=insertI(digits[j])
                        text1= ''.join(text1)
                        text.append(text1)
                j+=1
    else:
        text.append("the number is more 1000")
    text = ''.join(text)
    print(text)



#--------------------code for run-------------------------------------
try:
    #number =922
    number = int(input("Enter your number type int(0-1000):"))
    if number>=0 and number<=1000:
        number=int(number)
        print("your number is ",number)
        romantext(number)
    else:
        print("your number is not in range 0 to 1,000")
except Exception as e:
    print(e)

```
