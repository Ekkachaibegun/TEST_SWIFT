```python

"""
Convert Number to Thai Text.
เขียนโปรแกรมรับค่าจาก user เพื่อแปลง input ของ user ที่เป็นตัวเลข เป็นตัวหนังสือภาษาไทย
โดยที่ค่าที่รับต้องมีค่ามากกว่าหรือเท่ากับ 0 และน้อยกว่า 10 ล้าน

*** อนุญาตให้ใช้แค่ตัวแปรพื้นฐาน, built-in methods ของตัวแปรและ function พื้นฐานของ Python เท่านั้น
ห้ามใช้ Library อื่น ๆ ที่ต้อง import ในการทำงาน(ยกเว้น ใช้เพื่อการ test การทำงานของฟังก์ชัน).

"""
```

```python
#----------------------setting dictionary-----------------------------
my_dictionary1 = {0:"ศูนย์",1:"หนึ่ง",2:"สอง",3:"สาม",4:"สี่",5:"ห้า",6:"หก",7:"เจ็ด",8:"แปด",9:"เก้า"}
my_dictionary2 = {1:"เอ็ด",2:"ยี่"}
representing = {1:"",2:"สิบ",3:"ร้อย",4:"พัน",5:"หมื่น",6:"แสน",7:"ล้าน"}
# print(my_dictionary1.items())
# print(my_dictionary1.keys())
# print(my_dictionary1.values())
# print(my_dictionary1[0])





#----------------------function--------------------------------------
def thaitext(input_number,input_number2):
    text=[]
    if input_number<=9:#case
        text.append(my_dictionary1[input_number])
    elif input_number==10000000:#case
        text.append("สิบล้าน")
    else:#another case
        digits = [int(digit) for digit in str(input_number)]#split number
        j=0#index
        for i in range(len(str(input_number)),0,-1):#11111=>5loops(5,4,3,2,1)
            if digits[j]!=0:
                if i==2 and digits[-2]==2:#ยี่สิบ
                    text.append(my_dictionary2[digits[j]])
                    text.append(representing[i])
                elif i==2 and digits[-2]==1:#สิบ
                    text.append(representing[i])
                elif i==1 and digits[-1]==1 and digits[-2]!=0:#หน่วย
                    text.append(my_dictionary2[digits[j]])
                else:
                    text.append(my_dictionary1[digits[j]])
                    text.append(representing[i])
            else:
                pass
            j+=1



    if input_number2>0:
        text.append("จุด")
        digits = [int(digit) for digit in str(input_number2)]
        for i in range(len(str(input_number2))):
            text.append(my_dictionary1[digits[i]])

    text = ''.join(text)
    print(text)



#--------------------code for run-------------------------------------
try:
    number =10000000
    #number = input("Enter your number type int(0-10000000):")
    if number>=0 and number<=10000000:
        if '.' in str(number):
            integer_part, decimal_part = str(number).split('.')
            result = [int(integer_part), int(decimal_part)]
            print("your number is ",number)
            thaitext(result[0],result[1])
        else:
            number=int(number)
            print("your number is ",number)
            thaitext(number,0)
    else:
        print("your number is not in range 0 to 10,000,000")
except Exception as e:
    print(e)

```
